```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv('Employee.csv')
```

```python
# Data Exploration
print("Unique values in each feature:")
for column in df.columns:
    print(column, ":", df[column].unique())
    print("Length:", len(df[column]))
```

```python
print("\nStatistical Analysis:")
print(df.describe())
```

```python
# Renaming columns
df.rename(columns={'Company': 'company', 'Age': 'age', 'Salary': 'salary', 
                   'Place': 'place', 'Country': 'country', 'Gender': 'gender'}, inplace=True)
```

```python
# Data Cleaning
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Data Analysis
filtered_data = df[(df['age'] > 40) & (df['salary'] < 5000)]
```

```python
# Plot a chart with age and salary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_data, x='age', y='salary')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# Count the number of people from each place and represent it visually
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='place')
plt.title('Number of people from each place')
plt.xlabel('Place')
plt.ylabel('Count')
plt.show()
```
